```python
vuelos.rename(columns={'year':'anho', 
                       'month':'mes',
                       'day': 'dia',
                       'dep_time': 'hora_partida',
                       'dep_delay': 'retraso_partida',
                       'arr_time': 'hora_arribo',
                       'arr_delay': 'retraso_arribo',
                       'carrier': 'empresa', 
                       'talinum': 'id_avion', 
                       'flight': 'vuelo', 
                       'origin': 'origen', 
                       'dest': 'destino', 
                       'air_time': 'horas_vuelo', 
                       'distance': 'distancia', 
                       'hour': 'hora', 
                       'minute': 'minuto'
                      }, inplace=True)
vuelos.head()
```

### 3. Indicar cuales son las columnas que tiene al menos un dato perdido


```python
missing_data = vuelos.isnull()
missing_data.head()

for column in missing_data.columns.values.tolist():
    lost_records = len(vuelos[column]) - missing_data[column].value_counts()
    print("Columna", column)
    if lost_records[0] != 0:
        print(lost_records[0], "datos perdidos")
    else:
        print("Datos completos")
    print("")
    
print(vuelos.count())
```

### 4. Filtrar todos las filas donde el retraso de vuelo de partida y llegada es un dato perdido

```python
missing_data = vuelos[["retraso_partida", "retraso_arribo"]]
print(missing_data[missing_data["retraso_partida"].isnull() & missing_data["retraso_arribo"].isnull()])
```

### 5. Hacer una copia del dataframe vuelos , en la copia eliminar todos los datos perdidos, ¿Verifique si justifica eliminarlos?

```python
vuelos_copy = vuelos.copy()
print(vuelos_copy)

missing_data = vuelos_copy.isnull()
for column in missing_data.columns.values.tolist():
    vuelos_copy.dropna(subset=[column], axis=0, inplace = True)

print(vuelos_copy)
```

### 6. Hacer ua copia del dataframe vuelos, en la copia rellenar todo los datos perdidos con cero

```python
vuelos_copy = vuelos.copy()
print(vuelos_copy)

missing_data = vuelos_copy.isnull()
for column in missing_data.columns.values.tolist():
    vuelos_copy[column].replace(np.nan, 0, inplace=True)

print(vuelos_copy)
```

### 7. Cuantos datos perdidos hay en los retrasos de partida y llegada

```python
missing_data = vuelos.isnull()

for column in missing_data.columns.values.tolist():
    lost_records = len(vuelos[column]) - missing_data[column].value_counts()
    if column == "retraso_partida" or column == "retraso_arribo":
        print("Columna", column)
        if lost_records[0] != 0:
            print(lost_records[0], "datos perdidos")
        else:
            print("Datos completos")
```
