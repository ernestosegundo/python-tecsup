### Lectura de archivo CSV

```python
archivo_csv = "WEO_Data_Latinoamerica_2020_1.csv"
df = pd.read_csv(archivo_csv, thousands=',', decimal='.')
df.head()
```

```python
df.rename(columns={'Country': 'Pais'}, inplace=True)
df.head()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
# Caso 1
listado = ['Peru', 'Brazil', 'Chile']
anhos = list(map(str,range(2000, 2020))) # ['2000','2001',....,'2019']
# anhos = [str(i) for i in range(2000, 2020)] # ['2000','2001',....,'2019']
df_paises = df.loc[listado,anhos].transpose()
df_paises.describe()
```

```python
df_paises.plot(kind="box" )
#df_paises.plot(kind="box", figsize=(10, 7), color='blue', vert=False)
plt.title('Box plot : PPP of Perú-Chile-Brazil from 2000 - 2019')
plt.ylabel('Thousands millions($)')
plt.savefig("img/boxplot-csv")
plt.show()
```

Ejercicio : Calcular el diagrama de caja de Argentina y Colombia desde el año 2000 al 2015

```python
# Caso 2 : top 5 in 2019
nro_top=5
df_top = df.sort_values(["2019"], ascending = False).head(nro_top)
anhos2000s = list(map(str, range(2000, 2010)))
anhos2010s = list(map(str, range(2010, 2020)))
#print(anhos2000s)
#print(anhos2010s)
df_top.head()
```

```python
df_top_2000s = df_top.loc[:,anhos2000s].sum(axis=1)  # for column
df_top_2000s.head()
```

```python
df_top_2010s = df_top.loc[:,anhos2010s].sum(axis=1)  # for column
df_top_2010s.head()
```

```python
data = {"2000s":df_top_2000s, "2010s":df_top_2010s}
df_decades = pd.DataFrame(data)
df_decades.head()
```

```python
df_decades.plot(kind='box')
plt.title('PPP top 5 countries in Latinoamericana for decades 2000s and 2010s')
plt.show()
```

```python
# Q1 (25%)
# Q3 (75%)
# IQ = Q3 - Q1
# Outlier > Q3 + 1.5*IQ   
df_decades.describe()
```

```python
Q1 = df_decades.describe()["2000s"]['25%']
Q3 = df_decades.describe()["2000s"]['75%']
IQ = Q3 - Q1
threshold =  Q3 + 1.5*IQ

outliers = df_decades[df_decades['2000s'] > threshold]

print("Outliers : {}".format(outliers))
```

```python
# Caso 3 

listado = ['Peru','Brazil', 'Chile']
anhos = list(map(str,range(2000,2020)))
df_paises = df.loc[listado,anhos].transpose()

fig = plt.figure()
ax0 = fig.add_subplot(1,2,1) 
ax1 = fig.add_subplot(1,2,2) 

ax0.set_title("Box plot PPP : Perú-Chile-Brazil from 2000-2021")
df_paises.plot(kind="box", color='blue', vert=True, figsize=(10, 6), ax = ax0)

ax1.set_title("Line PPP : Perú-Chile-Brazil from 2000-2021")
df_paises.plot(kind="line", figsize=(10, 6), ax = ax1 )

plt.show()
```
