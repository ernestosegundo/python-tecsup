```python
# 
import numpy as np  
import pandas as pd 

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
```

```python
# open the file and read it into a variable alice_novel
quijote_novel = open('el_quijote.txt', 'r', encoding="utf8").read()
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
quijote_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
quijote_wc.generate(quijote_novel)
```

```python
# display the word cloud
plt.imshow(quijote_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(quijote_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('y') # add the words said to stopwords
stopwords.add('que') # add the words said to stopwords
stopwords.add('de') # add the words said to stopwords
stopwords.add('lo') # add the words said to stopwords
stopwords.add('la') # add the words said to stopwords


# re-generate the word cloud
quijote_wc.generate(quijote_novel)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(quijote_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image # converting images into arrays

# save mask to alice_mask
quijote_mask = np.array(Image.open('silueta_quijote.jpeg'))
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(quijote_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
quijote_wc = WordCloud(background_color='white', max_words=2000, mask=quijote_mask, stopwords=stopwords)

# generate the word cloud
quijote_wc.generate(quijote_novel)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(quijote_wc, interpolation='bilinear')
plt.axis('off')
plt.savefig("quijote_wc")
plt.show()
```
