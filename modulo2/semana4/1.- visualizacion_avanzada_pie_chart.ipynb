## 1.- Pie Chart

https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

#### Cargando las librerias de Matplotlib

```python
# Usando "inline backend"
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
'''Cantidad de moviles vendidos en 2019'''
etiquetas = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'HMD Global','Otros']
valores = [296.194, 240.615, 193.475, 126.049, 118.693, 565.630]
colores = ['green', 'red', 'blue', 'purple','yellow', 'orange']
plt.pie(valores, labels=etiquetas, colors=colores)
plt.axis('equal')
plt.savefig("img/pie-01")
plt.show()
```

```python
'Cantidad de moviles vendidos hasta agosto del 2019'
etiquetas = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'HMD Global','Otros']
valores = [296.194, 240.615, 193.475, 126.049, 118.693, 565.630]
colores = [ 'green', 'red', 'blue', 'purple','yellow', 'orange']
explode = [0.1, 0, 0, 0, 0, 0.1]
plt.title('Participación en las ventas del año 2019')
plt.pie(valores, labels=etiquetas, colors=colores, explode=explode, startangle=90)
plt.axis('equal')
plt.savefig("img/pie-02")
plt.show()
```

```python
'Cantidad de moviles vendidos hasta agosto del 2019'
font = {'family': 'serif',
        'color': 'red',
        'weight': 'normal',
        'size': 16,
        }
etiquetas = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'HMD Global','Otros']
valores = [296.194, 240.615, 193.475, 126.049, 118.693, 565.630]
colores = ['green', 'red', 'blue', 'purple','yellow', 'orange']
explode = [0.1, 0, 0, 0, 0, 0]
plt.title('Participación en las ventas del año 2019', fontdict=font)
plt.pie(valores, labels=etiquetas, colors=colores, explode=explode, shadow=True,autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.savefig("img/pie-03")
plt.show()
```

### Gráficos con Pandas

```python
import pandas as pd

data = {'ventas-2019': [296.194, 240.615, 193.475, 126.049, 118.693, 565.630]}
df = pd.DataFrame(data)
df['ventas-2019'].plot(kind='pie', figsize=(10, 5))
plt.axis('equal')
plt.savefig("img/pie-04")
plt.show()
```

```python
data = {'ventas-2019': [296.194, 240.615, 193.475, 126.049, 118.693, 565.630]}
df = pd.DataFrame(data)
colores = [ 'green', 'red', 'blue', 'purple','yellow', 'orange']
df['ventas-2019'].plot(kind='pie',colors=colores, figsize=(10, 5))
plt.savefig("img/pie-05")
plt.show()
```

```python
font = {'family': 'serif',
        'color': 'red',
        'weight': 'normal',
        'size': 16,
        }
colores = [ 'green', 'red', 'blue', 'purple','yellow', 'orange']
explode = [0.1, 0, 0, 0, 0, 0]
data = {'ventas-2019': [296.194, 240.615, 193.475, 126.049, 118.693, 565.630]}
etiquetas = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'OPPO','Otros']
df = pd.DataFrame(data, index=etiquetas)
df['ventas-2019'].plot(kind='pie',colors= colores, explode=explode,
                   shadow=True,autopct='%1.1f%%', startangle=90, figsize=(10, 5))
plt.title('Participación en las ventas del año 2019',fontdict=font)
plt.ylabel("") # evita que salga "ventas-2019"
plt.savefig("img/pie-06")
plt.show()
```

```python
archivo_csv = "WEO_Data_Latinoamerica_2020_1.csv"
df = pd.read_csv(archivo_csv, thousands=',', decimal='.')
df.rename(columns = {'Country': 'Pais'}, inplace=True)
df.set_index('Pais', inplace=True)
listado = ['Argentina','Peru','Brazil', 'Chile']
df_paises = df.loc[listado, '2019']
df_paises.plot(kind="pie", figsize=(12, 6))
plt.legend(listado,loc="upper left")
plt.legend(loc="upper left")
plt.ylabel("") # evita que salga el 2019
plt.savefig("img/pie-07")
plt.show()
```
